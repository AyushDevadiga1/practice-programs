```python
import numpy as np
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 
```

```python
df=pd.read_csv('L1.csv')
df.info()
df.head()
df.describe()
```

```python
plt.scatter(x=df.SAT,y=df.GPA)
plt.xlabel('SAT score')
plt.ylabel('GPA score')
plt.title('SAT vs GPA ')
plt.show()
```

```python
X=df[['SAT']]
y=df[['GPA']]
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
X_train , X_test , y_train , y_test = train_test_split(X, y , test_size=0.2 , random_state=25 )
#The random state was originall 42 I changed it to 25 to check R**2 value 
```

```python
model=LinearRegression()
model.fit(X_train,y_train)
```

```python
print("Slope (m):", model.coef_[0])
print("Intercept (b):", model.intercept_)
```

```python
print("The R^2 value of the model is :", model.score(X_test,y_test))
```

```python
plt.scatter(X,y,color='black')
plt.plot(X,model.predict(X),color='red')
plt.xlabel('SAT score')
plt.ylabel('GPA score')
plt.title('Linear Regression : SAT vs GPA ')
plt.show()
```
