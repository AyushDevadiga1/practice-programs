```python
class Cyphertext:

    def __init__(self,text = 'AXYnsndfn' , key = 3 ):
        self.text = text 
        self.key = key
        self.encrypted_message = None

    def encryption(self):

        encrypted_text = ''

        for i in self.text:
            if i.isupper():
                encrypted_text += chr(ord(i)+self.key)
            else:
                encrypted_text += chr(ord(i)+self.key)
        self.encrypted_message = encrypted_text

    def validate(self):
        if self.text.isalpha():
            self.encryption()
        else:
            print('Incorrect format ! Try Again')


    def get_encrypted_message(self):
        return self.encrypted_message
```
