```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
scalar_1 = [ 1, 5, -2 ]
# A set of scalars
```

```python
# A set of vectors
v1 = np.array([1,5,8])
v2 = np.array([5,2,6])
v3 = np.array([5,2,7])
```

```python
# To perform : we have to calculate the linear weighted combo
# first combine the vectors to a matrix
```

```python
matrix_a = [ v1 , v2 , v3]
```

```python
lin_wght_combo = np.zeros(len(v1))
```

```python
for i,j in zip(matrix_a,scalar_1):
    lin_wght_combo += i*j
```
